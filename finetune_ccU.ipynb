```python
# When the model fine-tuned, you receive mail from openai. In message you see id your model.
model_id = "ft:gpt-3.5-turbo-0613:personal::85vHgnGz"
```

```python
# Now I want to estimate fine-tuned model. I am going to ask simple questions from dataset 
# like 'What is Article 55 of the Criminal Code of Ukraine about?'
res = {}

for i in range(1, 10):
    num = f"{i+50}"
    mes = f"Про що стаття стаття {num} кримінального кодексу України?" # What about
    
    messages = [
        {"role": "system", "content": "Ти викладач права"},
        {"role": "user", "content": mes}]
    
    answer = openai.ChatCompletion.create(
        model=model_id, 
        messages=messages,
        temperature=0.5)
    
    res[num] = answer['choices'][0]['message']['content']
```

```python
for i in res:
    print(i, '\n\n', res[i])
```
